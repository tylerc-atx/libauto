```python
import re
h = ["https://192.168.0.10:5000/",
     "https://192.168.0.10:5000",
     "https://192.168.0.10/",
     "https://192.168.0.10",

     "http://192.168.0.10:5000/",
     "http://192.168.0.10:5000",
     "http://192.168.0.10/",
     "http://192.168.0.10",

     "https://192.168.0.10:5000/page.html",
     "https://192.168.0.10/page.html",

     "http://192.168.0.10:5000/page.html",
     "http://192.168.0.10/page.html",
    ]
```

```python
exp = "https?://([^:/]+)"

for t in h:
    print(re.search(exp, t).group(1))
```
